```python
while True:
    n = input()

    if not n:
        continue

    if not n.isdigit(): #??? 이해 ok, 다른 방법?
        continue

    N = int(n)
    if 1 <= N <= 20:
        break
```

```python
categories = [
    ("PAYMENT", ["pay", "card", "refund", "price"]),
    ("ACCOUNT", ["login", "password", "signup", "account"]),
    ("BUG", ["error", "bug", "crash", "fail"])
]

cnt_payment = 0
cnt_account = 0
cnt_bug = 0
cnt_other = 0

results = []
```

```python
for _ in range(N):
    text = input()
    text = text.strip()
    
    if not text:
        continue

    lower_text = text.lower()

    label = "OTHER"
    found = False
```

```python
    for category_name, keywords in categories:
        for key in keywords:    #내부 for문 어려움
            if key in lower_text:
                label = category_name
                found = True
                break
        
        if found:
            break
```

```python
    if label == "PAYMENT":
        cnt_payment +=1
    elif label == "ACCOUNT":
        cnt_account +=1
    elif label == "BUG":
        cnt_bug +=1
    else:
        cnt_other += 1
    
    results.append(f"[{label}] {lower_text}")

for r in results:
    print(r)

print(f"PAYMENT:{cnt_payment} ACCOUNT:{cnt_account} BUG:{cnt_bug} OTHER:{cnt_other}")

#??? 원하는 결과 느낌 아님, 못찾겠음
```
